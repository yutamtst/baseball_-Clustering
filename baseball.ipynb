```
df = pd.read_csv('pithcer.csv')
```

```
df.drop(df.columns[0],axis=1,inplace=True)
```

```
df
```

```
print(df.columns)
```

```
an_index = ['順位','被安打','与四球','自責点','奪三振']
```

```
an = df[an_index]
```

```
an
```

```
df['投球回']
```

```
for y in range(len(df['投球回'])):
  for x in range(len(an.columns)-1):
    an.iloc[y,x+1]=an.iloc[y,x+1]/df['投球回'][y]
```

```
an
```

```
import japanize_matplotlib
import seaborn as sns 
sns.set(font='IPAexGothic')
```

```
an.describe()
```

```
import matplotlib.pyplot as plt
from pandas import plotting 

plotting.scatter_matrix(an[an.columns[1:]],figsize=(10,10),alpha=0.8,diagonal='kde')
plt.show()
```

```
from sklearn.cluster import KMeans
kmeans_model = KMeans(n_clusters = 4, random_state=10).fit(an.iloc[:,1:])
```

```
labels = kmeans_model.labels_
```

```
labels
```

```
color_codes = {0:'red',1:'blue',2:'yellow',3:'green'}
```

```
colors =[color_codes[x]for x in labels]
```

```
plotting.scatter_matrix(an[an.columns[1:]],figsize=(10,10),color=colors,alpha=0.8,diagonal='kde')
plt.show()
```

```
from sklearn.decomposition import PCA
pca = PCA()
pca.fit(an.iloc[:,1:])
```

```
feature = pca.transform(an.iloc[:,1:])
```

```
df
```

```
plt.figure(figsize=(15,15))
for x,y,name in zip(feature[:,0],feature[:,1],an.iloc[:,0]):
  plt.text(x,y,name,alpha=0.8,size =10)
plt.scatter(feature[:,0],feature[:,1],alpha=0.8,color=colors)
plt.title('PCA Pithcer')
plt.xlabel('The first principal component score')
plt.ylabel('The second principal component score')
plt.show()
```
